```python
import pandas as pd # convention is to include pd 'alias' for reference purposes 
import numpy as np
```

```python
undata_df = pd.read_excel("UNData3.xlsx") 
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
g = sns.swarmplot(x="CONTINENT", y="LifeExp", data=undata_df)
for item in g.get_xticklabels():item.set_rotation(45)
g.set(xlabel = 'Continent', ylabel = 'Life Expectancy')
```
